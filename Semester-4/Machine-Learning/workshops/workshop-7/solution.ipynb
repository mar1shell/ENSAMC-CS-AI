# Les arbres de régression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
x = np.sort(5 * np.random.rand(200), axis=0)
y = np.sin(x)
```

```python
plt.plot(x, y, "r.")
plt.title("courbe de repartition sinusoidale")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
X_train = x.reshape(-1, 1)
reg = DecisionTreeRegressor(max_depth=2)
reg.fit(X_train, y)
X_test = np.sort(5 * np.random.rand(80, 1), axis=0)
y_predicted = reg.predict(X_test)
y_test = np.sin(X_test)
plt.figure()
plt.scatter(x, y, color="b", label="Exemples d'apprentissage")
plt.plot(X_test, y_predicted, color="r", label="Prédiction", linewidth=3)
plt.xlabel("x");    plt.ylabel("y")
plt.title("Régression par un arbre de décision")
plt.legend(); plt.show()
a = mean_squared_error(y_test, y_predicted)
print("mean_squared_error:", a)
```
