# Rapport TP4 Text Mining

Catégorisation d'un corpus avec **TF-IDF** et **K-Means**.

## Méthode

- Corpus: 18 documents (thèmes principaux: chats et services Google/Chrome).
- Vectorisation: `TfidfVectorizer(stop_words='english')`.
- Clustering: K-Means testé pour `K=2` à `K=10`.
- Évaluation: Inertie (Elbow) + score de Silhouette.

## Résultats

- La Silhouette maximale est obtenue pour **K = 6** (≈ 0.0399).
- L'inertie décroît régulièrement quand K augmente.
- Les clusters séparent globalement les contenus *chat* et *Google/Chrome*, avec des sous-thèmes pour K plus élevé.

## Conclusion

La valeur recommandée est **K = 6**, car elle offre le meilleur compromis observé (Silhouette la plus élevée et segmentation plus informative).
