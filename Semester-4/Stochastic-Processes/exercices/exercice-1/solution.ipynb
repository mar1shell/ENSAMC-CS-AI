## Exercice 4 — Loi de Bernoulli et Binomiale

### Partie 1 : Variable aléatoire de Bernoulli

Soit $X$ une variable aléatoire de loi de Bernoulli de paramètre $p$.

#### 1) Proposer une fonction qui construit une réalisation de $X$.

**Réponse :**

```python
def bernoulli(p):
    """
    Génère une réalisation d'une variable de Bernoulli de paramètre p
    
    Args:
        p: probabilité de succès (0 <= p <= 1)
    
    Returns:
        0 ou 1
    """
    return 1 if npr.rand() < p else 0
```

#### 2) Proposer une fonction qui renvoie un $k$-échantillon de cette variable aléatoire.

**Réponse :**

```python
def bernoulli_sample(p, k):
    """
    Génère un k-échantillon d'une variable de Bernoulli de paramètre p
    
    Args:
        p: probabilité de succès (0 <= p <= 1)
        k: taille de l'échantillon
    
    Returns:
        array de k réalisations
    """
    
    return [bernoulli(p) for _ in range(k)]
```

#### 3) Tester ces résultats en Python pour différentes valeurs de $p$ et $k$.

**Réponse :**

```python
print(bernoulli_sample(0.3, 10))
print(bernoulli_sample(0.5, 20))
```

### Partie 2 : Variable aléatoire Binomiale

Simuler une variable aléatoire $X$ de loi binomiale de paramètres $(n, p)$ pour différentes valeurs de $n$ et $p$.

**Réponse :** 

```python
def binomial_simulation(n, p):
    """
    Simule une variable aléatoire de loi binomiale B(n, p)
    
    Args:
        n: nombre d'essais
        p: probabilité de succès
    
    Returns:
        Une réalisation de la loi binomiale
    """
    return sum(bernoulli_sample(p, n))
```

```python
print(binomial_simulation(10, 0.3))
print(binomial_simulation(20, 0.5))
```

## Exercice 5 — Loi Géométrique

Soit $X$ une variable aléatoire géométrique de paramètre $p$. Proposer un algorithme pour la simulation de cette variable aléatoire.

#### 1) En utilisant le résultat : Si $(Y_i)_{i \in \mathbb{N}^*}$ sont des variables aléatoires i.i.d. de loi de Bernoulli de paramètre $p$, alors $N = \min\{i : Y_i = 1\}$ suit une loi géométrique de paramètre $p$.

**Réponse :**

```python
def geometric_method1(p):
    """
    Simule une loi géométrique de paramètre p (méthode 1)
    En utilisant des variables de Bernoulli
    
    Args:
        p: probabilité de succès (0 < p <= 1)
    
    Returns:
        N: min{i : Y_i = 1}
    """
    n = 1
    
    while True:
        if bernoulli_sample(p, 1)[0] == 1:
            return n
        n += 1
```

#### 2) En utilisant le résultat : Si $U$ suit la loi uniforme sur $[0, 1]$, alors $G = 1 + \left\lfloor \frac{\ln(U)}{\ln(1-p)} \right\rfloor$ est une variable aléatoire qui suit la loi géométrique de paramètre $p$.

**Réponse :**

```python
def geometric_method2(p):
    """
    Simule une loi géométrique de paramètre p (méthode 2)
    En utilisant la loi uniforme
    
    Args:
        p: probabilité de succès (0 < p <= 1)
    
    Returns:
        G: variable géométrique
    """
    return 1 + np.floor(np.log(npr.rand()) / np.log(1 - p))
```

#### 3) Comparer les résultats en insistant sur le temps de calcul.

**Réponse :**

```python
def compare_methods(p, n):
    # Méthode 1
    start_time = time.time()
    samples1 = [geometric_method1(p) for _ in range(n)]
    time1 = time.time() - start_time

    # Méthode 2
    start_time = time.time()
    samples2 = [geometric_method2(p) for _ in range(n)]
    time2 = time.time() - start_time

    print(f"Temps d'exécution méthode 1: {time1:.4f} secondes")
    print(f"Temps d'exécution méthode 2: {time2:.4f} secondes")
```

```python
compare_methods(0.3, 10000)
```
