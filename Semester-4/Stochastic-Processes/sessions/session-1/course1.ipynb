## Uniform Law simulation:

```python
import math
import random
```

```python
def uniform(n):
    u = random.random()
    return math.floor(u * n) + 1

def simulate_die_rolls(num_rolls):
    rolls = []
    
    for _ in range(num_rolls):
        roll = uniform(6)
        rolls.append(roll)
    
    return rolls 
```

```python
print(simulate_die_rolls(10))
```

## Bernoulli Law Simulation:

```python
def bernoulli(p):
    u = random.random()
    
    return 1 if u < p else 0

def simulate_bernoulli_trials(k, p):
    return [bernoulli(p) for _ in range(k)]

```

```python
print(simulate_bernoulli_trials)
```

## Uniform 2:

```python
def uniform2(a, b):
    return a + (b - a) * random.random()

def simulate_uniform_trials(k, a, b):
    return [uniform2(a, b) for _ in range(k)]
```

## BOX MULLER:

```python
def boxmuller():
    r = math.sqrt(-2 * math.log(random.random()))
    t = 2 * math.pi * random.random()
    return r * math.cos(t), r * math.sin(t)

def simulate_normal_trials(k, mu, sigma):
    return [mu + sigma * boxmuller()[0] for _ in range(k)]
```

```python
print(simulate_normal_trials(10, 0, 1))
```

```python
import matplotlib.pyplot as plt

data = simulate_normal_trials(1000, 0, 1)
plt.hist(data, bins=30, density=True)
plt.title('Histogram of Simulated Normal Trials')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

```

## Loi de Cauchy:

```python
def cauchy():
    u = random.random()
    return math.tan(math.pi * (u - 0.5))

def simulate_cauchy_trials(num):
    return [cauchy() for _ in range(num)]
```

```python
plt.hist(simulate_cauchy_trials(1000), bins=30, density=True)
plt.title('Histogram of Simulated Cauchy Trials')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

## Loi de Weibbull:

```python
def weibull(a, lambda_val):
    u = random.random()
    return lambda_val * (-math.log(u)) ** (1 / a)


def simulate_weibull_trials(num, k, lambda_val):
    return [weibull(k, lambda_val) for _ in range(num)]
    
```

```python
n, bins, patches = plt.hist(simulate_weibull_trials(1000, 2, 1), bins=30, density=True)
plt.title('Histogram of Simulated Weibull Trials')
plt.xlabel('Value')
plt.ylabel('Density')

bin_centers = (bins[:-1] + bins[1:]) / 2
plt.plot(bin_centers, n, 'r-', linewidth=2, label='Peak line')
plt.legend()
plt.show()
plt.hist(simulate_weibull_trials(1000, 2, 1), bins=30, density=True)
plt.title('Histogram of Simulated Weibull Trials')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
